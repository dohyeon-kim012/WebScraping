## 데이터 확인 및 테스트 크롤링

```python
# 'duckduckgo' 사이트의 Netword-XHR 에서 RequestURL을 복사
url = 'https://duckduckgo.com/i.js?l=wt-wt&o=json&q=10%20year%20face&vqd=3-123897281933951978045102314746858131276-291233585751825993113979507856558647155&f=,,,,,&p=1'

user_agent = {'User-agent' : 'Mozilla/5.0'}
response = requests.get(url, headers = user_agent)
print(response)
```

### 데이터 확인

```python
# 1. json 방식으로 로드하여 데이터 확인
import json
response_content = response.content 

response_json = json.loads(response_content)
response_json
```

```python
# 2. BeautifulSoup 방식으로 로드하여 데이터 확인
response_content = response.content 

response_soup = BeautifulSoup(response_content, 'html.parser')
response_soup
```

### 테스트 크롤링

```python
# json 방식으로 사진 url 만 불러와 리스트에 저장해보기
face_10_list = []

for idx in range(len(response_json['results'])):

    face_10 = response_json['results'][idx]['image']
    face_10_list.append(face_10)

face_10_list[:10]
```

# 크롤링

```python
from urllib.request import urlretrieve
import requests
from bs4 import BeautifulSoup

import json
import os
from tqdm.notebook import tqdm

url_teen = 'https://duckduckgo.com/i.js?l=wt-wt&o=json&q=teenager%20face&vqd=3-280589695270496546883710029147047398085-291233585751825993113979507856558647155&f=,,,,,&p={}'
url_10 = 'https://duckduckgo.com/i.js?l=wt-wt&o=json&q=10%20year%20face&vqd=3-123897281933951978045102314746858131276-291233585751825993113979507856558647155&f=,,,,,&p={}'
url_20 = 'https://duckduckgo.com/i.js?l=wt-wt&o=json&q=20%20year%20face&vqd=3-197762750866313343659250251810854016360-291233585751825993113979507856558647155&f=,,,,,&p={}'
url_30 = 'https://duckduckgo.com/i.js?l=wt-wt&o=json&q=30%20year%20face&vqd=3-41122823152535309200575417878444420169-291233585751825993113979507856558647155&f=,,,,,&p={}'
url_50 = 'https://duckduckgo.com/i.js?l=wt-wt&o=json&q=50%20year%20face&vqd=3-236183244382029138143669385136236556084-291233585751825993113979507856558647155&f=,,,,,&p={}'
url_60 = 'https://duckduckgo.com/i.js?l=wt-wt&o=json&q=60%20year%20face&vqd=3-120953603933134881691651960082845303213-291233585751825993113979507856558647155&f=,,,,,&p={}'
url_70 = 'https://duckduckgo.com/i.js?l=wt-wt&o=json&q=70%20year%20face&vqd=3-81336141689212619047378055034341082122-291233585751825993113979507856558647155&f=,,,,,&p={}'

user_agent = {'User-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

face_60_list = []

for page in tqdm(range(1, 5)):
    url_final = url_60.format(page)
    response = requests.get(url_final, headers = user_agent)
    response_content = response.content
    response_json = json.loads(response_content)

    # dir_name = "./{}".format(response_content['query'])
    # os.mkdir(dir_name)

    for idx in range(len(response_json['results'])):
        face_60 = response_json['results'][idx]['image']
        face_60_list.append(face_60)

        for face_60_img in face_60_list:
            try:
                filename = response_json['results'][idx]['title']
                urlretrieve(face_60_img, "/content/drive/MyDrive/실습_도현,원식,현균/dataset/image_added/2/{}".format(filename + '.jpg'))
            except:
                pass
```
